# Experimental Hypothesis
    
<p style="font-family:times">
    
The U. S Food and Nutrition Service (FNS) encompasses 15 federal nutrition assistant programs, the aid ranges from providing food grants to low-income to middle class Women, Infants, and Children (WIC) to Summer Food programs. 

This data set represents users of 5 out of the 15 programs in the U.S states for the years 2009,2011-2016.  As observed in the pairplot, the number of users is directly correlated to the population size and to each other. 

For this proposal, the year 2015 is selected based on being the most recent year with data for all the programs. In addition, two states with similar population size and similar number of users for each food program are selected in order to run the A/B experiment. 

As observed bellow, Illinois and Pennsylvania show the National School Lunch Program with the highest number of users followed by School Breakfast, Women Infant and Child, Child and Adult Care, and the Summer Food respectively. The large discrepancy between the Summer Food program to the rest could be attributed to those food programs being offered in schools during the regular annual school days.

The summer food program feeds only 15 children for every 100 children that use the program during regular school day. This is mostly attributed to a reduction of places that offer the program, some participating institutions donâ€™t stay open the entire summer, and limited transportation. 

Therefore, testing an implementation of transportation options similar to those offered during regular school days is worth the while if less kids go hungry. The food program is sponsored by the U. S Department of Agriculture (USDA) and partly by the Department of Education, some of the budget utilize would be allocated to put some school buses in circulation and pick up children to bring them to the programs. The null hypothesis states that the implementation of transportation aid would not have an effect on an increase of users of the summer food program. The experimental hypothesis states that overcoming the road blocks of transportation during the summer by providing transportation to children with none to limited access to the food program, will significantly show an increase of users.
</p>

# Rollout
<p style="font-family:times">

Transportation â€“ School buses will be used so they are easily recognized by the children and it is familiar to them as a safe way of transportation to have their breakfast or lunches. 

Counties per State â€“ Ten counties, in each state, will be chosen at random to reduces the probability that the effects of the experiment are due to the countyâ€™s demography or income instead of the effect of transportation. However, at least five counties in each state had to have shown a reduction of users at some point in the past six years.

Time frame of experiment â€“ Transportation would be offered for half of the summer, if there is an increase of about 0.10% of users (the percentage of users over population), the same program would be implemented to the rest of the counties in each state for the second half of the summer. If the increase is less than 0.10% the transportation would be offered to another additional ten random counties in each state applying the same conditions shown above. 

Note that between 2009-2015 (no data for 2010), Pennsylvania had its highest participation rate for the Summer Program in 2013 with 0.81% and Illinois had the lowest participation rate of 0.39% in 2009. Both states have roughly the same population but in the last six years Pennsylvania, at some point, had twice the participants in the Summer Program compared to Illinois. Therefore, it is known that a participation rate of 0.81% is possible. 
</p>

# Evaluation of Success
<p style="font-family:times">
    
The evaluation key is to run a t-test with 0.5 degrees of freedom between ten random counties in which the transportation was not provided as the control group and ten counties that had buses offering transportation to the places offering the Summer Program. The t-test will also be performed for the second half of the summer with also 0.5 degrees of freedom. 

The success of the experiment will be measure as a statistically significant increase of users getting to the Summer Program in either the first or second half of the program. 

In order to account for a possible effect of  the location of counties selected in either half section of the summer, or number of counties, or time that takes for the transportation to show effect, the t-test with a statistical significance of 95% is accepted for either half of the summer as a measurement of success of the experiment.
</p>
