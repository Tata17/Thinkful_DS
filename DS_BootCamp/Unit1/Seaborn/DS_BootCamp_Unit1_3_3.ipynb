## CHALLENGE

The following graphs show the 2015 breakfast, lunch, and summer food programs in the U.S for each state (states represented by a number). In addition to the 2015 population of each state.

The goal is to see how the different continuos variables such as number of people reached by the food programs and the population number vary across states (categorical variable).


```python
df3 = df2[['StateFIPS', 'National_School_Lunch_Program_participants_FY_2015', 
          'School_Breakfast_Program_participants_FY_2015', 'Summer_Food_participants_FY_2015',
          'State_Population_2015']]
df3.columns = ['States', 'Lunch15', 'Breakfast15', 'Summer15', 'Pop15']
df3.head()
```

```python
# slice out the categorical variable for now.
ax = sns.pairplot(df3, vars= df3.columns[1:])
```

As seen above, the numbers are big and it skews the data, specially the histograms. To have a better representation of the data, lets take a look at its logarithmic scale. Note this practice has to be used with care, know your data. 

```python
# Take the log of population and gdp_per_capita
df3['logLunch15'] = np.log10(df3['Lunch15'])
df3['logBreakfast15'] = np.log10(df3['Breakfast15'])
df3['logSummer15'] = np.log10(df3['Summer15'])
df3['logPop15'] = np.log10(df3['Pop15'])
```

```python
# Drop the non-transformed columns
df3 = df3.drop(columns = ['Lunch15', 'Breakfast15', 'Summer15', 'Pop15'])

df3.head()
```

```python
ax= sns.pairplot(df3, vars= df3.columns[1:])
```

We can see a cleaner distribution of the data, although the outliers are not as highlighted. The number one general conclusion we can derive is that all the food programs are highly correlated to each other and to population number. We see a slightly higher concentration of points where we observe the highest population numbers.

```python
ax = sns.pairplot(df3, vars= df3.columns[1:], hue='States')
```

The graph above shows the different programs distribution by state across the U.S. If we refer to the names of the states for the corresponding numbers, we could pin point the states that show outliers in population size for 2015 as those that have more or less people enroll in their food program.
